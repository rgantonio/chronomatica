# Importing Packages

```python
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageFilter
```

# Loading Dataset

```python
# Load CIFAR-100 dataset without any transformation
original_dataset = torchvision.datasets.CIFAR100(root='./data', train=True, download=True)
image, label = original_dataset[2]  # Get first image
```

# Apply Transforms

```python
resize_transform = transforms.Resize((224, 224))
resized_image = resize_transform(image)
# Apply sharpening
sharpened_image = resized_image.filter(ImageFilter.SHARPEN)
```

# Display

```python
def show_image_variants(original_img, resized_img, sharpened_img):
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))

    axs[0].imshow(original_img)
    axs[0].set_title("Original (32x32)")
    axs[0].axis('off')

    axs[1].imshow(resized_img)
    axs[1].set_title("Resized (224x224)")
    axs[1].axis('off')

    axs[2].imshow(sharpened_img)
    axs[2].set_title("Sharpened (224x224)")
    axs[2].axis('off')

    plt.tight_layout()
    plt.show()

show_image_variants(image, resized_image, sharpened_image)
```
